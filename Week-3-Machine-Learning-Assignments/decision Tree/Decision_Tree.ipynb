```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from  sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('50_Startups.csv')
# X = df.iloc[:, :-1].values
# y = df.iloc[:, 4].values
# y = y.reshape(-1, 1)
# y
independent = df[['R&D Spend', 'Administration', 'Marketing Spend']]
dependent = df['Profit']
```

```python
train_test_split(independent, dependent, test_size=0.2, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.2, random_state=0)
```

```python
df.columns
```

```python
standard_scaler = StandardScaler()
X_train = standard_scaler.fit_transform(X_train)
X_test = standard_scaler.transform(X_test)
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train) 
y_pred = regressor.predict(X_test)
accuracy = r2_score(y_test, y_pred)
accuracy
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```
