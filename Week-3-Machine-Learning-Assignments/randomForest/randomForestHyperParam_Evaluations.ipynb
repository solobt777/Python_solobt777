```python
def run_test_scenario(X_train, X_test, y_train, y_test, scenario):
    test_id = scenario.pop('test_id')
    purpose = scenario.pop('purpose')

    #print(f"\nRunning {test_id}: {purpose}")
    print(f"Parameters: {test_id}: {scenario}")

    # Create and train model
    model = RandomForestRegressor(**scenario)
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    results = {
        'test_id': test_id,
        'purpose': purpose,
        'mse': mse,
        'rmse': rmse,
        'mae': mae,
        'r2': r2,
        'feature_importances': model.feature_importances_
    }
    
    if scenario.get('oob_score'):
        results['oob_score'] = model.oob_score_
    
    print(f"Results - RMSE: {rmse:.4f}, MAE: {mae:.4f}, R²: {r2:.4f}")
    
    return results
```

```python
# Load your data

# Run all test scenarios
results = []

for scenario in test_scenarios[19:30]:
    result = run_test_scenario(X_train, X_test, y_train, y_test, scenario.copy())
    results.append(result)

# Create results DataFrame
#results_df = pd.DataFrame(results)
#print(results_df)
# test = results_df['feature_importances']
# for n in test:
#     print(n)
```
