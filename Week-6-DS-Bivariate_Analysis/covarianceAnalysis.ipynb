```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

df = pd.read_csv('Placement.csv')
df.head()
```

```python
# fillna values in salary column
# The correct way to check for NaN values in a pandas Series is using .isna() or .isnull()
df = df.fillna(df['salary'].isna())
df = df.fillna(df['salary'].mean())
df.isna().sum()
```

```python
dir(df)
df.select_dtypes(include=[np.number]).cov()
```

```python
# covariance between two columns 'degree_p' and 'etest_p' and explain what covariance  indicate whether is is positive or negative
covariance = df['degree_p'].cov(df['etest_p'])
print("Covariance between degree_p and etest_p are postitively:", covariance)

# covariance between two columns 'etest_p' and 'mba_p' and explain what covariance  indicate whether is is positive or negative
covariance = df['etest_p'].cov(df['mba_p'])
print("Covariance between etest_p and mba_p are postitively:", covariance)
```

## Corelationship

```python
df.select_dtypes(include=[np.number]).corr()
```

## Positive Correlation r=1
- Perfect positive Correlation (r =1)
- Highest degree of +ve Correlation (r = High or 0.95)
- Low degree of +ve Correlation (r = + low or 0.54)


```python
# positive correlation between ssc_p vs hsc_p and printing the result
correlation = df['ssc_p'].corr(df['hsc_p'])
print("correlation between ssc_p and hsc_p are LOW DEGREE CORRELATIONSHIP:", correlation)
```

## Negative Correlation (r = -1)
- Perfect positive Correlation (r = -1)
- Highest degree of +ve Correlation (r = -low or -0.54)
- Low degree of +ve Correlation (r = - high or -0.94)

```python
# NEGATIVE correlation between degree_p vs salary and printing the result
correlation = df['degree_p'].corr(df['salary'])
print("correlation between degree_p and salary are HIGHEST DEGREE OF CORRELATIONSHIP:", correlation)
```
