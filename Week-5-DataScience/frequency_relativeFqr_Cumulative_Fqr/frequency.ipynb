```python
def frequencyTable(column_name, dataset):
    
    #for col in column_name:
        frequencyTable = pd.DataFrame(columns=["UniqueValues", "Frequency", "RelativeFrequency", "Cumsum"])
        #print(dataset[col].value_counts().index)
        frequencyTable["UniqueValues"] = dataset[column_name].value_counts().index
        frequencyTable["Frequency"] = dataset[column_name].value_counts().values
        frequencyTable["RelativeFrequency"] = (frequencyTable["Frequency"] / len(dataset[column_name]))
        frequencyTable["Cumsum"] = frequencyTable["RelativeFrequency"].cumsum()
        return frequencyTable
```
