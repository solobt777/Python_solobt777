## how to fill the entier dataset with missing value in it

```python
df = pd.read_csv('Placement.csv')
df.head()
```

```python
from Univariate import UnivariateCls
quant, qual = UnivariateCls(df).quanQual()
print(quant)
print(qual)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
# Get column indices for quant columns
quant_indices = [i for i in range(len(quant))]

imputer.fit(df[:, quant_indices])
df[:, quant_indices] = imputer.transform(df[:, quant_indices])
df[:, quant_indices]

df.head()
```
