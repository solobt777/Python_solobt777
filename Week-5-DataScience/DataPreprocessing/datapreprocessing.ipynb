```python
# write a function to get_pdf_probability passing dataset for a specific column, start_range, end_range and plot the pdf
def get_pdf_probability(data, column, start_range, end_range):
    mean = data[column].mean()
    std = data[column].std()
    x = np.linspace(mean - 4*std, mean + 4*std, 1000)
    pdf = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std) ** 2)
    
    plt.plot(x, pdf, label='PDF')
    plt.fill_between(x, pdf, where=(x >= start_range) & (x <= end_range), color='gray', alpha=0.5, label='Probability Area')
    plt.xlabel(column)
    plt.ylabel('Probability Density')
    plt.title(f'PDF of {column} with Probability Area from {start_range} to {end_range}')
    plt.legend()
    plt.show()
    
    probability = np.trapz(pdf[(x >= start_range) & (x <= end_range)], x[(x >= start_range) & (x <= end_range)])
    return probability
```
