## converting the normal distribution to Standard Normal Distribution

- Standard Normal Deviation means
-3  -2  -1  0   1   2   3

- Always Standard Normal Distribution
mu = 0, and sd = 1

- to convert any standard deviation like e.g.(mu=150 & sd=2.4) or (mu=77.4 & sd =40.1) or (mu=20.33 & sd=5.65)
- so we have to convert such sd deviation deviating from in mu=0, sd=1 to normal distribution
- how to convert such standard normal distribution to NOrmal Distribution
- we can use a formula call Z-Score Z = x-mu/sd -> this is also called as standard scalar

## Below is the way to converting normal distribution to Standard Normal Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Placement.csv')
df.head()
```

```python
df_salary_empty = df.fillna(df['salary'].isna())
df_salary_empty = df.fillna(df['salary'].mean())
df_salary_empty.isna().sum()
```

```python
def standNormalize(df):
    mean = df.mean()
    std = df.std()
    values = [value for value in df]
    
    z_Scores = [(value - mean) / std for value in values]
    
    sns.distplot(z_Scores, kde=True)
    plt.title('Standard Normal Distribution')
    plt.xlabel('Z-Scores')
    plt.ylabel('Frequency')
    plt.show()
```

```python
standNormalize(df['salary'])
```
